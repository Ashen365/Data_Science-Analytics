```python
import pandas as pd 
from apyori import apriori #apyori=
```

```python
movie_data = pd.read_csv("movie_dataset.csv",header=None)
```

```python
movie_data
```

```python
display(movie_data.head())
```

```python
records = [] 
for i in range(0, 7501): 
    records.append([str(movie_data.values[i, j]) for j in range(0, 20)])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.5,min_length=2) 
#list of list
```

```python
association_results = list(association_rules)
```
